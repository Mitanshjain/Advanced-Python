```python
# Encoding --- > Transform your categorical data into Numerical data .
# (1). LabelEncoder -- > 1-1 column apply but we cannot set the order manually .
# (2). OrdinalEncoder -- > multi-column apply and we can set the order manually .
# (3). OneHotEncoder --- > has_covid[yes,no]--> 2 column create --> covid_yes, covid_no
# (4). get_dummies --- > has_covid[yes,no]---> 2 columns create ---> column name with values
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/covid_toy - covid_toy.csv")
```

```python
df = df.dropna()
```

```python
df.drop(columns = ['city'], inplace = True)
```

```python
# Apporach 1 ---> LabelEncoder
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
```

```python
df['cough'] = lb.fit_transform(df['cough'])
df['gender'] = lb.fit_transform(df['gender'])
df['has_covid'] = lb.fit_transform(df['has_covid'])
```

```python
df.head(2)
```

```python
# Apporach 3 ---> OneHotEncoder
```

```python
df = pd.read_csv("/content/covid_toy - covid_toy.csv")
df.head(2)
```

```python
df = df.dropna()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder( drop = 'first', sparse_output = False, dtype = np.int32)
```

```python
ohe.fit_transform(df[['gender','cough', 'city', 'has_covid']])
```

```python
# gender ---> f
# cough ---> s
# city ---> m,b,k
# # has_covid --> n
```

```python
# Apporach 4 ---> get.dummmies
```

```python
df = pd.read_csv("/content/covid_toy - covid_toy.csv")
df.head(2)
```

```python
df = df.dropna()
```

```python
df = pd.get_dummies(df, columns = ['gender','cough', 'city', 'has_covid'], drop_first=True)
```

```python
df = df.astype(int)
```
